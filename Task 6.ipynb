## Task - Create a Decision Tree Classifier on Iris data set , see prediction and visualize the tree

### collection necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.externals.six import StringIO  
from IPython.display import Image 
import pydotplus
```

### Data Collection, presentation and preparation

```python
raw_data = pd.read_csv(r'C:\Users\DEBOPAM DAS\Desktop\New folder\GRIP_may_21\Iris.csv')
```

```python
raw_data.head()
```

```python
X=raw_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y=raw_data['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.1,random_state=0)
```

```python
Y.value_counts()
```

### Classifier building and training

```python
Classifier=DecisionTreeClassifier()
Classifier.fit(x_train,y_train)
print("training complete...")
```

### Testing the classifier

```python
y_pred=Classifier.predict(x_test)
a=y_pred
b=y_test.values
c=pd.DataFrame({'Actual':b,'Predicted':a})
c
```

### Predicting and Vizualization

```python
a=[[4.3,3.5,2.3,2.0]]
b=Classifier.predict(a)
b
```

```python
dot_data = StringIO()
export_graphviz(Classifier, out_file=dot_data, feature_names=X.columns
               ,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
